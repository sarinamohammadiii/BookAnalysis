# 2.Reading data

```python
import pandas as pd
df = pd.read_csv("books_to_scrape.csv")
df.head()
```

# 3.EDA

```python
df.info()
df.describe()
df.isnull().sum()
```

# 4.Numerical and statistical analysis (book prices)

```python

mean_price = df["book_price"].mean()
min_price = df["book_price"].min()
max_price = df["book_price"].max()

print("mean_price:", mean_price)
print("min_price:", min_price)
print("max_price:", max_price)
```

# 4.Numerical and statistical analysis (book rating)

```python

mean_rating = df["book_rating"].mean()
min_rating = df["book_rating"].min()
max_rating = df["book_rating"].max()

print("mean_rating:", mean_rating)
print("min_rating:", min_rating)
print("max_rating:", max_rating)
```

# 4.Numerical and statistical analysis (book avail)

```python

mean_avail = df["book_avail"].mean()
min_avail = df["book_avail"].min()
max_avail = df["book_avail"].max()

print("mean_avail:", mean_avail)
print("min_avail:", min_avail)
print("max_avail:", max_avail)
```

# 5.Number of books in each category

```python
# 
category_counts = df["book_cat"].value_counts()
print("Number of books in each category:\n", category_counts)
```

# 6.Visualization (price histogram)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df["book_price"], bins=20, kde=True, color="skyblue")
plt.title("Distribution of book prices")
plt.xlabel("price")
plt.ylabel("Number of books")
plt.show()
```

# 6.Visualization(Number of books in each category)

```python

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,10))
sns.countplot(y="book_cat", data=df, order=df["book_cat"].value_counts().index, palette="viridis")
plt.title("Number of books in each category")
plt.xlabel("Number of books")
plt.ylabel("Book categories")
plt.show()
```

# 6.Visualization (The relationship between book ratings and prices)

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,5))
sns.scatterplot(x="book_rating", y="book_price", data=df, color="purple")
plt.title("The relationship between book ratings and prices")
plt.xlabel("price")
plt.ylabel("rating")
plt.show()
```

# 6.Visualization(Price per category)

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(25,10))
sns.boxplot(x="book_cat", y="book_price", data=df, palette="Set2")
plt.title("Book price distribution in each category")
plt.xlabel("Book categories")
plt.ylabel("pice")
plt.xticks(rotation=45) 
plt.show()
```

# 7.Advanced Analysis (Highest Rated Book)

```python

top_rated_books = df.sort_values(by="book_rating", ascending=False).head(5)
top_rated_books[["book_title", "book_cat", "book_price", "book_rating"]]
```

# 7.Advanced Analysis (Highest price Book)

```python
top_expensive_books = df.sort_values(by="book_price", ascending=False).head(5)
top_expensive_books[["book_title", "book_cat", "book_price", "book_rating"]]
```

# 7.Advanced Analysis(correlation matrix , heatmap)

```python
numeric_cols = ["book_price", "book_rating", "book_avail"]
correlation_matrix = df[numeric_cols].corr()
print("\ncorrelation matrix")
display(correlation_matrix)
```

# 7.Advanced Analysis(correlation matrix , heatmap)

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation matrix of numeric columns")
plt.show()
```
